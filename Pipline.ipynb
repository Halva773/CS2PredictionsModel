# Введение
#### Суть поекта реализовать мультифункциональную предсказательную модель на киберспортивные, спортивные матчи

В нашем случае, необходима модель классификации (выбор например: П1/П2). Так же в цели проекта входит предсказание других параметров, таких как количество раундов, или дуэль между игроками. 
Теститься модель будет на киберспортивной сцене CS2. Так же ключевыми направлениями в будущем будет Valorant и футбол.Ъ

# Выбор метрик модели

Изначально цель - предсказать победителя карт(ы).
Все данные, которые будут собираться и использоваться для этой модели

1. `Map`: Название карты (категориальная переменная)
2. `Kpr`: *Среднее количество убийств за раунд 
3. `Adr`: *Средний урон за раунд
4. `Multikills`: *Среднее количество раундов, в котором игрок убил 3+ противников
5. `Kd`: *Соотношение убийств и смертей
6. `Hard Clutches (1v3+)`: Количество сложных клатчей
7. `Light Clutches (1v1/1v2)`: Количество простых клачей
8. `Clutches - acting out on communication (2v3+/3v5)`: *Средне количество клатчей, которое взял игрок за игру
9. `Wins from the minority`: Процент побед в раундах, где энтри был проигран, но раунд выигран
10. `Defeats from the majority`: Процент побед в раундах, где энтри был выигран, но раунд проигран
11. `Entry rate`: Процент раундов, в которых команда находит ентри
12. `Double-Entry rate`: Процент раундов, в которых раунд начинался 5в3
13. `Double-Entry loss-rate`: Процент поражений из ситуации 5в3
14. `Double-Entry get rate`: Процент раундов, в которых раунд начинался 3в5
15. `Double-Entry get win-rate`: Процент побед из ситуации 3в5
16. `Nades damage`: Урон с гранат (молики/ХЕ) - урон за раунд
17. `T-side win %`: Винрейт команды за Т сторону
18. `CT-side win %`: Винрейт команды за КТ сторону
19. `Team Tier (обе команды)`: **Тир команды
20. `Current Form (last 10 against all)`: Винрейт за последнение 10 матчей против любых команд
21. `Current Form (last 5 against top-10)`: Винрейт за последние 5 матчей против команд из топ 10
22. `Map played last 90 days (against top20)`: Количество сыграных (всех) карт против разных тиров за последние 60 дней против топ-20
23. `Played this map lst 90 days`: Наигрыш данной карты за последние 90 дней
24. `Points of the defeated teams`: Средий рейтинг HLTV побеждённых команд за последние 90 дней 
25. `Points of undefeated teams`: Средний рейтинг HTTV команд, которым проиграли за последние 90 дней
26. `Stage of tournament`: Стадия турнира:
    - Open_qualifications
    - Close_qualifications
    - Group_stage
    - PlayOff
    - PlayOff_Last_chanse
27. `Relegation match`: Матч на вылет (True/Flase)
28. `Prise Pool (размер призового)`: Призовой фонд турнира
29. `Aces`: Количество эйсов игроков/карта
30. `Rating hltv`: Очки рейтинга HLTV 
31. `Pistol win %`: Процент выигранных пистолетных раундов
32. `Awp kills/rnd`: Убийств с АВП/раунд
33. `AVG win rounds in lost`: Количество выиграных раундов в проигрышных матчах
34. `AVG lost rounds in win`: Количество проигранных раундов в победных матчах
35. `Smokes usage`: Среднее количество смоков, использованое за раунд
36. `Smokes assist`: Процент ассистов, которые сделаны с помощью смоков
37. `Kills in smoke`: Количество убийств, при выходе противника в смок
38. `Die in smoke`: Количество смертей, при выходе в смок
39. `Flash usage`: Среднее количество флешек, использованых за раунд
40. `Flash assists`: Процент ассистов, которые сделаны с помощью флешек
41. `Days this roaster`: Дней в этом составе
42. `Days this backbone`: Дни этим косяком
43. `Wined pool this backbone`: Выигранные призовые за последний год
